ðŸ“š THEORY PART
1. Descriptive Statistics
Descriptive statistics summarize and describe the important features of a dataset.

They help us understand the central tendency (typical value) and variability (spread) of the data.

2. Measures of Central Tendency
These indicate where most values fall:

Mean: The average (sum of all values divided by number of values).

Median: The middle value when data is sorted.

Mode: The most frequently occurring value (not used here, but important).

3. Measures of Variability
These show how spread out the values are:

Minimum: Smallest value.

Maximum: Largest value.

Standard Deviation: How much the data deviates from the mean on average.

Percentile: Indicates the value below which a given percentage of observations fall (example: 25th percentile, 50th percentile, 75th percentile).

4. Grouping by Categorical Variables
A categorical variable (e.g., species) divides the data into groups.

We then calculate statistics separately for each group (for example, the mean sepal length for each species).
